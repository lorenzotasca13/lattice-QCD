```python
diff=np.zeros(100)
for i in np.arange(0,100):
    sigma=i*0.01
    x=sigma*np.random.randn(100000)+1
    diff[i]=np.std(x**2)-2*np.mean(x)*np.std(x)
plt.plot(np.arange(0,100), diff)
```
